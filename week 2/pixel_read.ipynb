```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
image = Image.open('test.bmp','r')
image_gray = image.convert("L")
```

```python
width, height = image.size
print("Image width:", width)
print("Image height:", height)
```

```python
# Display pixel values in matrix appearance
for y in range(height):
    for x in range(width):
        pixel_value = image.getpixel((x, y))
        print(pixel_value, end=" ")
    print()  # Move to the next line for the next row
```

```python
# Display the image
plt.imshow(image_gray, cmap='gray')
plt.title("Hover over the image to get pixel values")
plt.colorbar()

# Callback function to get pixel value on mouse hover
def on_hover(event):
    if event.xdata is not None and event.ydata is not None:
        x = int(event.xdata)
        y = int(event.ydata)
        pixel_value = image_gray.getpixel((x, y))
        print(f"Pixel value at ({x}, {y}): {pixel_value}")

plt.connect('motion_notify_event', on_hover)
plt.show()
```

```python
def crop_half_vertical(image):
    width, height = image.size
    cropped_image = image.crop((0, 0, width, height//2))
    cropped_image.show()
```

```python
crop_half_vertical(image)
```

```python
def crop_half_horizontal(image):
    width, height = image.size
    cropped_image = image.crop((0, 0, width//2, height))
    cropped_image.show()
```

```python
crop_half_horizontal(image)
```
