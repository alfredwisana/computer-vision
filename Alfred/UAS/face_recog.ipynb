```python
import face_recognition
import matplotlib.pyplot as plt
import cv2
```

```python
# Load the image file
image_path = "Multi_Racial_Face_he3-2321570-696x706.gif"


# Open the GIF file
capture = cv2.VideoCapture(image_path)

ret,frame = capture.read()
image_temp = frame.copy()
# image = face_recognition.load_image_file(image_path)

# Find all face locations in the image
face_locations = face_recognition.face_locations(frame)
face_landmarks_list = face_recognition.face_landmarks(frame)

print("Found {} faces in the image.".format(len(face_locations)))

# Display the image with face bounding boxes
for face_location in face_locations:
    top, right, bottom, left = face_location
    cv2.rectangle(image_temp, (left, top), (right, bottom), (255,0, 255), 2)


plt.figure(figsize=(10,6))

plt.subplot(1,2,1)
plt.imshow(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))
plt.title("Original Image")
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(cv2.cvtColor(image_temp, cv2.COLOR_BGR2RGB))
plt.title("Detected Face")
plt.axis('off')
plt.show()
```
