PENGURANGAN PIXEL

```python
new_pixel=[[0 for _ in range(height)] for _ in range(width)]
for y in range(height):
    for x in range(width):
        pixel_value = image.getpixel((x, y))
        if (pixel_value - 10 >= 0):
            pixel_value = pixel_value -10
            new_pixel[x][y]=pixel_value
        else:
            new_pixel[x][y]=0
        
```

```python
array = np.array(new_pixel, dtype=np.uint8)
array = np.rot90(array,k=3)
array = np.flip(array, axis=1)

new_image = Image.fromarray(array)
new_image.show()
```

MEMOTONG GAMBAR

```python
def crop_half_vertical(image):
    width, height = image.size
    cropped_image = image.crop((0, 0, width, height//2))
    cropped_image.show()
```

```python
crop_half_vertical(image)
```

```python
def crop_half_horizontal(image):
    width, height = image.size
    cropped_image = image.crop((0, 0, width//2, height))
    cropped_image.show()
```

```python
crop_half_horizontal(image)
```
