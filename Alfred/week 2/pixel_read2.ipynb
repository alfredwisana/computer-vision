```python
import cv2
import matplotlib.pyplot as plt
```

```python
# Load the image
image = cv2.imread('test.bmp', cv2.IMREAD_GRAYSCALE)
```

```python
# Print the width and height of the image
height, width = image.shape
print("Image width:", width)
print("Image height:", height)
```

```python
# Display pixel values in matrix appearance
for y in range(height):
    for x in range(width):
        pixel_value = image[y, x]
        print(pixel_value, end=" ")
    print()  # Move to the next line for the next row
```

```python
# Display the image using Matplotlib
plt.imshow(image,cmap='gray')

# Define a function to get pixel value on hover
def hover(event):
    if event.inaxes:
        x, y = int(event.xdata), int(event.ydata)
        if 0 <= x < image.shape[1] and 0 <= y < image.shape[0]:
            pixel_value = image[y, x]
            print(f"Pixel value at ({x}, {y}): {pixel_value}")

# Connect the hover event
plt.connect('motion_notify_event', hover)

# Show the plot
plt.show()
```
