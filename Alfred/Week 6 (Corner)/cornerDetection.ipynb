```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
from scipy import ndimage
```

```python
filename = 'triangle5.png'
image = cv.imread(filename)
image2 = ndimage.rotate(image, 123)

gray = cv.cvtColor(image,cv.COLOR_BGR2GRAY)
gray2 = cv.cvtColor(image2,cv.COLOR_BGR2GRAY)
```

```python
gray = np.float32(gray)
gray2 = np.float32(gray2)


corner = cv.cornerHarris(gray,2,5,0.04)
corner2 = cv.cornerHarris(gray2,2,5,0.04)
 
#result is dilated for marking the corners, not important
corner = cv.dilate(corner,None)
corner2 = cv.dilate(corner2,None) 

# Threshold for an optimal value, it may vary depending on the image.
threshold = corner > 0.01*corner.max()
threshold2 = corner2 > 0.01*corner2.max()


image[threshold]= [0,0,255]

image2[threshold2]=[0,0,255]
```

```python
num_corners = np.sum(image == [0,0,255])
print("Number Of Corner: %s"%num_corners)

plt.figure(figsize=(15,15))

plt.subplot(1,2,1)
plt.imshow(image,cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(corner)
plt.title('Corner')
plt.axis('off')


plt.show()
```
