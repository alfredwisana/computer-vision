```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
from scipy import ndimage
```

```python
def haris_corner(image):


    gray = cv.cvtColor(image,cv.COLOR_BGR2GRAY)
    gray = np.float32(gray)

    corner = cv.cornerHarris(gray,2,5,0.04)

    corner = cv.dilate(corner,None)

    threshold = corner > 0.01*corner.max()

    image[threshold]= [0,255,255]


    plt.figure()
    plt.imshow(image, cmap='gray')
    plt.title('Corner Detection')
    plt.axis('off')    

```

```python
def haris_corner_rotate(image, rotation_angle):
    
    image = ndimage.rotate(image, rotation_angle)

    haris_corner(image)

    
```

```python
filename = 'rectangle.png'
image = cv.imread(filename)

haris_corner(image)
```
